Types of Nave Bayes Models:

* Gaussian Nave Bayes

--> Used when the features are continuous and follow a normal (Gaussian) distribution.

--> Example: Predicting whether a student will pass/fail based on continuous features like study hours, marks, etc.

Multinomial Nave Bayes

--> Used for discrete features (counts).

--> Works well for text classification (spam detection, sentiment analysis).

--> Example: Word counts in documents.

Bernoulli Nave Bayes

--> Used for binary/boolean features (yes/no, 0/1).

--> Example: Whether an email contains a particular word (yes/no).

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
gnb=GaussianNB()
mnb=MultinomialNB()
bnb=BernoulliNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
mnb.fit(x_train,y_train)
```

```python
bnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test),gnb.score(x_test,y_test)
```

```python
mnb.score(x_test,y_test),mnb.score(x_test,y_test)
```

```python
bnb.score(x_test,y_test),bnb.score(x_test,y_test)
```

```python
plt.figure(figsize=(10,8))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
```
